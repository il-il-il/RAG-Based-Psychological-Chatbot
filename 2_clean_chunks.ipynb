```python
import pickle

# Load the previously saved documents
with open("/content/drive/MyDrive/clean_documents.pkl", "rb") as f:
    documents = pickle.load(f)
```

```python

# ✅ Split into chunks
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(documents)

```

```python
# Filter again after splitting (sometimes split creates blanks)
texts = [
    doc for doc in chunks
    if isinstance(doc, Document)
    and isinstance(doc.page_content, str)
    and doc.page_content.strip()
]

print(f"Total valid chunks: {len(texts)}")
```

```python
import pickle

with open("/content/drive/MyDrive/clean_chunks.pkl", "wb") as f:
    pickle.dump(texts, f)

print("Filtered chunks saved successfully.")
```
